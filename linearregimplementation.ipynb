```python
import math
import numpy as np
import pandas as pd
import plotly.express as px
import pickle
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv('study.csv')
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['Attendance_Hours'], df['Final_Marks'], color='red')
plt.xlabel('Attendance Hours')
plt.ylabel('Final Marks')
plt.show()
```

```python
def gradient_descent(m_now,b_now, data, L):
    n=int((len(data)))
    m_gradient=0
    b_gradient=0
    for i in range(n):
        x=data.Attendance_Hours[i]
        y=data.Final_Marks[i]
        m_gradient+=(-1)*(2/n)*(y-(m_now*x+b_now))*x
        b_gradient+=(-1)*(2/n)*(y-(m_now*x+b_now))
    
    m=m_now-(L*m_gradient)
    b=b_now-(L*b_gradient)
    
    return m,b
```

```python
m=0
b=30# trial and error
epochs=1000
learning_rate=0.0001

for i in range(epochs):
    m,b=gradient_descent(m,b,df,learning_rate)


print(m,b)
print("The best fitted line is y={}x+{}".format(m,b))


#plotting the obtained line against initial data
plt.figure(figsize=(10,6))
x=np.linspace(9,81,300)
y=m*x+b
plt.plot(x,y, color='blue', label='Fitted Line')
plt.scatter(df['Attendance_Hours'], df['Final_Marks'], color='red')
plt.xlabel('Attendance Hours')
plt.ylabel('Final Marks')
plt.show()
```
